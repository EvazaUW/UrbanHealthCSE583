Inorder to run app in jupyter notebook including all shiny set steps:
1. windows powershell open as administrator
conda install -c conda-forge shiny
2. create your shiny app in a separate file, e.g.:
  '''app.py''' \
  import shiny \
  from shinywidgets import output_widget, render_widget \
  app_ui = ui.page_fluid(
      ui.input_slider("n", "N", 0, 100, 10),
      output_widget("val")
  ) \
  def server(input, output, session):
      @render_widget
      def val():
          return ui.output_text_verbatim(f"N * 2 is {input.n() * 2}") \
  app = shiny.App(app_ui, server)
  
3. In shell run:
shiny run --port 8000 app.py

4. Start notebook and run:
from IPython.display import IFrame
IFrame(src='http://localhost:8000/', width=800, height=600)

```python
from shiny.express import input, ui
from shinywidgets import render_altair

ui.input_selectize("var", "Select variable", choices=["bill_length_mm", "body_mass_g"])


@render_altair
def hist():
    import altair as alt
    from palmerpenguins import load_penguins
    df = load_penguins()
    return (
        alt.Chart(df)
        .mark_bar()
        .encode(x=alt.X(f"{input.var()}:Q", bin=True), y="count()")
    )
```
